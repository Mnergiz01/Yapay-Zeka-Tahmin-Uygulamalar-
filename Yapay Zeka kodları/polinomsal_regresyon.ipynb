```python
import numpy as np
import matplotlib.pyplot as plt
x = [1,2,3,5,6,7,8,9,10,12,13,14,15,16,18,19,21,22,23,24,25]
y = [100,90,80,60,60,55,60,65,70,70,75,76,78,79,90,99,99,100,98,97,90]
#deney 1:
#verileri 3.derece polinom ile modelleyelim
polymodel = np.polyfit(x,y,3)
#modeli bir polinom fonksiyonuna dönüştürelim
pol=np.poly1d(polymodel)
#linspace ile 1 den 25 e kadar eşit aralıklı 100 değer oluşturalım
x1=np.linspace(1,25,100)
#modelin verileri ne ölcüde temsil ettiğini görelim
plt.scatter(x,y)
plt.plot(x1,pol(x1))
plt.show()
```

```python
#deney 2 :
#acaba bu model veri aralığı dışında nasıl sonuç verir
x1=np.linspace(-2,30,100)
plt.scatter(x,y)
plt.plot(x1,pol(x1))
plt.show()
```

```python
polymodel=np.polyfit(x,y,2)
pol = np.poly1d(polymodel)
# linspace ile -11 den 30 kadar eşit aralıklı 100 değerli oluşturalım
x1=np.linspace(-2,30,100)

# Sonuç biraz aşırı genelleme gibi değilmi? Model karmaşıklığı artırılmalı
plt.scatter(x, y)
plt.plot(x1, pol(x1))
plt.show()
```

```python
# Deney 3: Model karmaşıklığını yükseltirseniz ne olabilir?
# bunun için verileri 20. derece polinom ile modolleyelim.
polymodel=np.polyfit(x, y, 20)
# Modeli bir polinom fonksiyonuna dönüştürelim
pol = np.poly1d(polymodel)
# linspace ile -11 den 30 kadar eşit aralıklı 100 değerli oluşturalım
x1 = np.linspace(1, 25, 100)
# Aşırı öğrenmeye yakın bir sonuç değil mi?
# Model veri noktalarını çok iyi temsil ediyor. Nerede ise aynı değeri veriyor.
# Veri noktalarına çok yakın olmasına rağmen bazı bölgelerde ciddi sapmalar olduğunu farkediniz.
# Bu bölgelerde (diğer bir ifade ile eğitim kümesinde kullanılmayan veri bazı verilerde) önemli modelleme hatasına yol açabilir.
plt.scatter(x, y)
plt.plot(x1, pol(x1))
plt.show()
```

```python
# Deney 3: Model karmaşıklığını çok yükseltirseniz ne olabilir?
# bunun için verileri 50. derece polinom ile modolleyelim.
polymodel=np.polyfit(x, y, 50)
# Modeli bir polinom fonksiyonuna dönüştürelim
pol = np.poly1d(polymodel)
# linspace ile -11 den 30 kadar eşit aralıklı 100 değerli oluşturalım
x1 = np.linspace(1, 25, 100)
# Aşırı öğrenmeye yakın bir sonuç gibi.
# Model veri noktaları dışında ciddi modelleme hatası var. 10^6 mertebelerinde
plt.scatter(x, y)
plt.plot(x1, pol(x1))
plt.show()
```
