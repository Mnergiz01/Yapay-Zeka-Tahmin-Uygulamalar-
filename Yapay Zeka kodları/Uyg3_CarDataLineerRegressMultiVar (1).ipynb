```python
# Tablo ve veriler ile çalışmak için pandas import edildi
import pandas as pd
from sklearn import linear_model
# regresyon performansı hesaplamaları için
from sklearn.metrics import mean_absolute_error, mean_squared_error
import matplotlib.pyplot as plt

# CarData veri kümesinde araba ağırlıkları, motor hacimi ve CO2 salınım
# verileri var. CarData.csv dosyası Colab sol menüde Dosyalar kısmına yüklenmeli
data=pd.read_csv("CarData.csv")
print(data)

# Model yardımı ile ağırlık ve motor hacmi verilerine göre CO2 salınımını
# tahmin eden lineer regresyon modeli elde edelim.
X = data[['Volume', 'Weight']]
y = data['CO2']

# regresyon modeli oluşturulması
regr_model = linear_model.LinearRegression()
regr_model.fit(X, y)

#Ağırlığı 2300kg, motor hacmi is 1300cm3 için CO2 salınımını tahmin edelim
predictedCO2 = regr_model.predict([[2300, 1300]])

print('CO2 salınımı :',predictedCO2)

```

```python
# Deney 1: Bütün veri kümesi için tahminleri inceleyelim.
#Bütün eğitim veri kümesi için tahmin yapalım.
predictedCO2 = regr_model.predict(X)

# Model performansını değerlendirelim
mae = mean_absolute_error(y, predictedCO2)
mse = mean_squared_error(y, predictedCO2)
print(f'Ortalama Mutlak Hata (MAE): {mae:.4f}')
print(f'Ortalama Kare Hata (MSE): {mse:.4f}')

# Modelin tahminlerinin gerçek değerler ile birlikte inceleyelim.
#
plt.plot(y,'-o')
plt.plot(predictedCO2)
plt.xlabel('Data No')
plt.ylabel('CO2 Tahminleri')
plt.show()
```
